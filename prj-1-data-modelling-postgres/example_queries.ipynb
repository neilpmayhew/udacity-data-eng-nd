# Most popular artists

```python
%%sql

SELECT a.name artist_name,COUNT(*) NumberOfPlays
FROM songplays sp
JOIN artists a
ON a.artist_id = sp.artist_id
GROUP BY a.name
ORDER BY NumberOfPlays DESC
```

# Most common week day and hour to list

```python
%%sql

SELECT t.weekday, t.hour,COUNT(*) NumberOfPlays
FROM songplays sp
JOIN time t
ON t.start_time = sp.start_time
GROUP BY t.weekday, t.hour
ORDER BY NumberOfPlays DESC
```
